```python
import pandas as pd
import numpy as np
df = pd.read_csv("../Data/train.csv")
```

```python
def feature_engineering(df_train):
  b1 = df_train["b1"]
  b2 = df_train["b2"]
  b3 = df_train["b3"]
  b4 = df_train["b4"]
  b5 = df_train["b5"]
  b6 = df_train["b6"]
  b7 = df_train["b7"]
  b8 = df_train["b8"]
  b8a = df_train["b8_a"]
  b9 = df_train["b9"]
  b11 = df_train["b11"]
  b12 = df_train["b12"]

  df_train['NDVI'] = (df_train['b8'] - df_train['b4']) / (df_train['b8'] + df_train['b4'])
  df_train['Adjusted transformed soil-adjusted VI'] = 1.22 * (b8 - 1.22 * b4 - 0.03) / (1.22 * b8 + b4 - 1.22 * 0.03 + 0.08 * (1 + 1.22 ** 2))
  df_train['Aerosol free vegetation index 1600'] = b8 - 0.66 * b11 / (b8 + 0.66 * b11)
  df_train['Aerosol free vegetation index 2100'] = b8 - 0.5 * b12 / (b8 + 0.56 * b12)
  df_train['Alteration'] = b11 / b12
  df_train['Anthocyanin reflectance index'] = 1 / b3 - 1 / b5
  # df_train['Ashburn Vegetation Index']
  # df_train['Atmospherically Resistant Vegetation Index']
  df_train['Atmospherically Resistant Vegetation Index 2'] = -0.18 + 1.17 * (b8 - b4) / (b8 + b4)
  df_train['Blue-wide dynamic range vegetation index'] = (0.1 * b8 - b2) / (0.1 * b8 + b2)
  df_train['Browning Reflectance Index'] = (1 / b3 - 1 / b5) / b8
  # df_train['Canopy Chlorophyll Content Index']
  df_train['Chlorophyll Absorption Ratio Index'] = (b5 / b4) * np.sqrt(((b5 - b3) / 150 * 670 + b4 + (b3 - ((b5 - b3) / 150 * 550))) ** 2) / np.sqrt(((b5 - b3) / (150 ** 2) + 1))
  # df_train['Chlorophyll Absorption Ratio Index 2']
  df_train['Chlorophyll Green'] = (b7 / b3) ** -1
  df_train['Chlorophyll Index Green'] = (b8 / b3) - 1
  df_train['Chlorophyll IndexRedEdge'] = (b8 / b5) - 1
  df_train['Chlorophyll Red-Edge'] = (b7 / b5) ** -1
  df_train['Chlorophyll vegetation index'] = b8 * b4 / (b3 ** 2)
  df_train['Coloration Index'] = (b4 - b2) / b4
  # df_train['Corrected Transformed Vegetation Index']
  df_train['CRI550'] = (b2.astype(float) ** -1) - (b3.astype(float) ** -1)
  df_train['CRI700'] = (b2.astype(float) ** -1) - (b5.astype(float) ** -1)
  df_train['Datt1'] = (b8 - b5) / (b8 - b4)
  df_train['Datt4'] = b4 / (b3 * b5)
  df_train['Datt6'] = b8a / (b3 * b5)
  df_train['Difference 678/500'] = b4 - b2
  df_train['Difference 800/550'] = b8 - b3
  df_train['Difference 800/680'] = b8 - b4
  df_train['Difference 833/658'] = b8 - b4
  df_train['Difference NIR/Green'] = b8 - b3
  #df_train['Differenced Vegetation Index MSS']
  df_train['EVI'] = 2.5 * (b8 - b4) / (b8 + (6 * b4) + (-7.5 * b2) + 1)
  df_train['EVI 2'] = 2.4 * (b8 - b4) / (b8 + b4 + 1)
  # df_train['EVI 2']
  # df_train['EPI']
  # df_train['Ferric iron, Fe2+']
  # df_train['Ferric iron, Fe3+']
  # df_train['Ferric Oxides']
  # df_train['Ferrous iron']
  df_train['Ferrous Silicates'] = b12 / b11
  df_train['Global Environment Monitoring Index'] =  (2 * (b8 ** 2 - b4 ** 2) + 1.5 * b8 + 0.5 * b4) / (b8 + b4 + 0.5) * (1 - 0.25 * (2 * (b8 ** 2 - b4 ** 2) + 1.5 * b8 + 0.5 * b4) / (b8 + b4 + 0.5)) - ((b4 - 0.125) / (1 - b4))
  #df_train['Global Vegetation Moisture Index']
  df_train['Gossan'] = b11 / b4
  df_train['Green atmospherically resistant vegetation index'] = (b8 - (b3 - (b2 - b4)))/(b8 - (b3 + (b2 - b4)))
  df_train['Green leaf index'] = (2 * b3 - b4 - b2) / (2 * b3 + b4 + b2)
  df_train['Green Normalized Difference Vegetation Index'] = (b8 - b3) / (b8 + b3)
  # df_train['Green Optimized Soil Adjusted Vegetation Index']
  df_train['Green Soil Adjusted Vegetation Index'] = (b8 - b3) / (b8 + b3 + 0.48) * (1 + 0.48)
  df_train['Green-Blue NDVI'] = (b8 - (b3 + b2)) / (b8 + (b3 + b2))
  df_train['Green-Red NDVI'] = (b8 - (b3 + b4)) / (b8 + (b3 + b4))
  df_train['Hue'] = np.arctan((2 * b4 - b3 - b2) / (30.5 * (b3 - b2)))
  # df_train['Ideal vegetation index']
  df_train['Infrared percentage vegetation index'] = (b8 / (b8 + b4)) / 2 * (df_train['NDVI'] + 1)
  df_train['Intensity'] = (1/30.5) * (b4 + b3 + b2)
  df_train['Inverse reflectance 550'] = b3.astype(float) ** -1
  df_train['Inverse reflectance 700'] = b5.astype(float) ** -1
  df_train['Laterite'] = b11 / b12
  df_train['Leaf Chlorophyll Index'] = (b8 - b5) / (b8 + b4)
  # df_train['Leaf Water Content Index']
  df_train['Log Ratio'] = np.log(b8 / b4)
  df_train['Maccioni'] = (b7 - b5) / (b7 - b4)
  df_train['MCARI/MTVI2'] = (((b5 - b4) - 0.2 * (b5 - b3)) * (b5 / b4)) / (1.5 * (1.2 * (b8 - b3) - 2.5 * (b4 - b3)) / (np.sqrt((2 * b8 + 1) ** 2 - (6 * b8 - 5 * np.sqrt(4)) - 0.5)))
  df_train['MCARI/OSAVI'] = ((b5 - b4) - 0.2 * (b5 - b3) * (b5 / b4))/ ((1 + 0.16) * (b8 - b4) / (b8 + b4 + 0.16))
  df_train['mCRIG'] = (b2.astype(float) ** -1 - b3.astype(float) ** -1) * b8
  df_train['mCRIRE'] = (b2.astype(float) ** -1 - b5.astype(float) ** -1) * b8
  # df_train['Mid-infrared vegetation index']
  # df_train['Misra Green Vegetation Index']
  # df_train['Misra Non Such Index']
  # df_train['Misra Soil Brightness Index']
  # df_train['Misra Yellow Vegetation Index']
  df_train['mND680'] = (b8 - b4) / (b8 + b4 - 2 * b1)
  df_train['Modified anthocyanin reflectance index'] = (b3.astype(float) ** -1 - b5.astype(float) ** -1) * b8
  df_train['Modified Chlorophyll Absorption in Reflectance Index'] = ((b5 - b4) - 0.2 * (b5 - b3)) * (b5 / b4)
  df_train['Modified Chlorophyll Absorption in Reflectance Index 1'] = 1.2 * (2.5 * (b8 - b4) - 1.3 * (b8 - b3))
  df_train['Modified Chlorophyll Absorption in Reflectance Index 2'] = 1.5 * (2.5 * (b8 - b4) - 1.3 * (b8 - b3)) / (np.sqrt((2 * b8 + 1) ** 2 - (6 * b8 - 5 * np.sqrt(b4)) - 0.5))
  df_train['Modified NDVI'] = (b8 - b4) / (b8 + b4 - 2 * b1)
  df_train['Modified Simple Ratio'] = (b8 - b1) / (b4 - b1)
  df_train['Modified Simple Ratio 670,800'] = ((b8 / b4) - 1) /np.sqrt((b8 - b4) + 1)
  df_train['Modified Simple Ratio NIR/RED'] = ((b8 / b4) - 1) /np.sqrt((b8 / b4) + 1)
  df_train['Modified Soil Adjusted Vegetation Index'] = (2 * b8 + 1 - np.sqrt((2 * b8 + 1) ** 2 - 8 * (b8 - b4)))/2
  df_train['Modified Soil Adjusted Vegetation Index hyper'] = (0.5) * ((2 * b8 + 1) - np.sqrt((2 * b8 + 1) ** 2 - 8 * (b8 - b4)))
  df_train['Modified Triangular Vegetation Index 1'] = 1.2 * (1.2 * (b8 - b3) - 2.5 * (b4 - b3))
  df_train['Modified Triangular Vegetation Index 2'] = 1.5 * (1.2 * (b8 - b3) - 2.5 * (b4 - b3)) / np.sqrt((2 * b8 + 1) ** 2 - (6 * b8 - 5 * np.sqrt(b4)) - 0.5)
  # df_train['Nonlinear vegetation index']
  df_train['Norm G'] = b3 / (b8 + b4 + b3)
  df_train['Norm NIR'] = b8 / (b8 + b4 + b3)
  df_train['Norm R'] = b4 / (b8 + b4 + b3)
  df_train['Normalized Difference 550/450'] = (b3 - b1) / (b3 + b1)
  df_train['Normalized Difference 550/650'] = (b3 - b4) / (b3 + b4)
  df_train['Normalized Difference 774/677'] = (b7 - b4) / (b7 + b4)
  df_train['Normalized Difference 780/550'] = (b7 - b3) / (b7 + b3)
  df_train['Normalized Difference 782/666'] = (b7 - b4) / (b7 + b4)
  df_train['Normalized Difference 790/670'] = (b7 - b4) / (b7 + b4)
  df_train['Normalized Difference 800/2170'] = (b8 - b12) / (b8 + b12)
  df_train['Normalized Difference 800/470'] = (b8 - b2) / (b8 + b2)
  df_train['Normalized Difference 800/500'] = (b8 - b2) / (b8 + b2)
  df_train['Normalized Difference 800/550'] = (b8 - b3) / (b8 + b3)
  df_train['Normalized Difference 800/650'] = (b8 - b4) / (b8 + b4)
  df_train['Normalized Difference 800/675'] = (b8 - b4) / (b8 + b4)
  df_train['Normalized Difference 800/680'] = (b8 - b4) / (b8 + b4)
  df_train['Normalized Difference 819/1600'] = (b8 - b11) / (b8 + b11)
  df_train['Normalized Difference 819/1649'] = (b8 - b11) / (b8 + b11)
  df_train['Normalized Difference 820/1600'] = (b8 - b11) / (b8 + b11)
  df_train['Normalized Difference 827/668'] = (b8 - b4) / (b8 + b4)
  df_train['Normalized Difference 833/1649'] = (b8 - b11) / (b8 + b11)
  df_train['Normalized Difference 833/658'] = (b8 - b4) / (b8 + b4)
  df_train['Normalized Difference 860/1640'] = (b8a - b11) / (b8a + b11)
  df_train['Normalized Difference 895/675'] = (b8 - b4) / (b8 + b4)
  df_train['Normalized Difference Green/Red'] = (b3 - b4) / (b3 + b4)
  # df_train['Normalized Difference MIR/NIR']
  df_train['Normalized Difference NIR/Blue'] = (b8 - b2) / (b8 + b2)
  df_train['Normalized Difference NIR/Green'] = (b8 - b3) / (b8 + b3)
  # df_train['Normalized Difference NIR/MIR']
  df_train['Normalized Difference NIR/Red'] = (b8 - b4) / (b8 + b4)
  df_train['Normalized Difference NIR/Rededge'] = (b8 - b5) / (b8 + b5)
  # df_train['Normalized Difference NIR/SWIR']
  df_train['Normalized Difference Red/Green'] = (b4 - b3) / (b4 + b3)
  df_train['Normalized Difference Salinity Index'] = (b11 - b12) / (b11 + b12)
  df_train['Normalized Difference Vegetation Index 690-710'] = (b8 - b5) / (b8 + b5)
  # df_train['Normalized Difference Vegetation Index C']
  df_train['Optimized Soil Adjusted Vegetation Index'] = (1 + 0.16) * (b8 - b4) / (b8 + b4 + 0.16)
  df_train['Pan NDVI'] = (b8 - (b3 + b4 + b2)) / (b8 + (b3 + b4 + b2))
  # df_train['Perpendicular Vegetation Index']
  # df_train['Ratio Analysis of Reflectance Spectra A1']
  # df_train['Ratio Analysis of Reflectance Spectra A2']
  # df_train['Ratio Analysis of Reflectance Spectra A3']
  # df_train['Ratio Analysis of Reflectance Spectra A4']
  # df_train['Ratio Analysis of Reflectance Spectra C3']
  # df_train['Ratio Analysis of Reflectance Spectra C4']
  df_train['RDVI'] = (b8 - b4) / ((b8 + b4).pow(0.5))
  df_train['RDVI2'] = (b8 - b4) / ((b8 + b4).pow(0.5))
  df_train['Red edge 1'] = b5 / b4
  df_train['Red edge 2'] = (b5 - b4) / (b5 + b4)
  df_train['Red-Blue NDVI'] = (b8 - (b4 + b2)) / (b8 + (b4 + b2))
  df_train['Red-Edge Inflection Point 1'] = 700 + 40 * ((((b4 + b7) / 2) - b5) / (b6 - b5))
  df_train['Red-Edge Inflection Point 2'] = 702 + 40 * ((((b4 + b7) / 2) - b5) / (b6 - b5))
  df_train['Red-Edge Inflection Point 3'] = 705 + 35 * ((((b4 + b7) / 2) - b5) / (b6 - b5))
  df_train['Red-Edge Position Linear Interpolation'] = 700 + 40 * ((((b4 + b7) / 2) - b5) / (b6 - b5))
  # df_train['Reduced Simple Ratio']
  df_train['Reflectance at the inflexion point'] = (b4 + b7) / 2
  df_train['Renormalized Difference Vegetation Index'] = (b8 - b4) / ((b8 + b4).pow(0.5))
  # df_train['SAVImir']
  df_train['Shape Index'] = (2 * b4 - b3 - b2) / (b3 - b2)
  df_train['Simple Ratio 1600/820'] = b11 / b8
  df_train['Simple Ratio 1650/2218'] = b11 / b12
  df_train['Simple Ratio 440/740'] = b1 / b6
  df_train['Simple Ratio 450/550'] = b1 / b3
  df_train['Simple Ratio 520/670'] = b2 / b4
  df_train['Simple Ratio 550/800'] = b3 / b8
  df_train['Simple Ratio 560/658'] = b3 / b4
  df_train['Simple Ratio 675/555'] = b4 / b3
  df_train['Simple Ratio 675/705'] = b4 / b5
  df_train['Simple Ratio 700'] = 1 / b5
  df_train['Simple Ratio 710/670'] = b5 / b4
  df_train['Simple Ratio 735/710'] = b6 / b5
  df_train['Simple Ratio 774/677'] = b7 / b4
  df_train['Simple Ratio 800/2170'] = b8 / b12
  df_train['Simple Ratio 800/500'] = b8 / b2
  df_train['Simple Ratio 810/560'] = b8 / b3
  df_train['Simple Ratio 833/1649'] = b8 / b11
  df_train['Simple Ratio 833/658'] = b8 / b4
  df_train['Simple Ratio 850/710'] = b8 / b5
  df_train['Simple Ratio 860/550'] = b8a / b3
  df_train['Simple Ratio 860/708'] = b8a / b5
  # df_train['Simple Ratio MIR/NIR']
  # df_train['Simple Ratio MIR/Red']
  df_train['Simple Ratio NIR/700-715'] = b8 / b5
  df_train['Simple Ratio NIR/G'] = b8 / b3
  # df_train['Simple Ratio NIR/MIR']
  df_train['Simple Ratio NIR/RED'] = b8 / b4
  df_train['Simple Ratio NIR/Rededge'] = b8 / b5
  df_train['Simple Ratio Red/Blue'] = b4 / b2
  df_train['Simple Ratio Red/Green'] = b4 / b3
  df_train['Simple Ratio Red/NIR'] = b4 / b8
  # df_train['Simple Ratio SWIRI/NIR']
  #Ignore 193 - 218
  df_train['Soil Adjusted Vegetation Index'] = (b8 - b4) / (b8 + b4 + 0.48) * (1 + 0.48)
  # df_train['Soil and Atmospherically Resistant Vegetation Index']
  # df_train['Soil and Atmospherically Resistant Vegetation Index 2']
  # df_train['Soil and Atmospherically Resistant Vegetation Index 3']
  # df_train['Soil Background Line']
  # df_train['Soil Composition Index']
  # df_train['Soil-adjusted vegetation index 2']
  # df_train['Specific Leaf Area Vegetation Index']
  df_train['SQRT(IR/R)'] = (b8 / b4).pow(0.5)
  df_train['Structure Intensive Pigment Index 1'] = (b8 - b1) / (b8 - b4)
  df_train['Structure Intensive Pigment Index 3'] = (b8 - b2) / (b8 - b4)
  # df_train['Tasselled Cap - brightness']
  # df_train['Tasselled Cap - Green Vegetation Index MSS']
  # df_train['Tasselled Cap - Non Such Index MSS']
  # df_train['Tasselled Cap - Soil Brightness Index MSS']
  # df_train['Tasselled Cap - vegetation']
  # df_train['Tasselled Cap - wetness']
  # df_train['Tasselled Cap - Yellow Vegetation Index MSS']
  df_train['TCARI/OSAVI'] = (3 * ((b5 - b4) - (0.2 * (b5 - b3) * (b5 / b4)))) / ((1 + 0.16) * ((b8 - b4)/(b8 + b4 + 0.16)))
  df_train['Transformed Chlorophyll Absorbtion Ratio'] = 3 * ((b5 - b4) - (0.2 * (b5 - b3) * (b5 / b4)))
  df_train['Transformed NDVI'] = ((b8 - b4) / (b8 + b4) + 0.5).pow(0.5)
  # df_train['Transformed Soil Adjusted Vegetation Index']
  # df_train['Transformed Soil Adjusted Vegetation Index 2']
  # df_train['Transformed Vegetation Index']
  df_train['Triangular chlorophyll index'] = 1.2 * (b5 - b3) - 1.5 * (b4 - b3) * (b5 / b4).pow(0.5)
  df_train['Vegetation Index 700'] = (b5 - b4) / (b5 + b4)
  df_train['Visible Atmospherically Resistant Index Green'] = (b3 - b4) / (b3 + b4 + b2)
  # df_train['Visible Atmospherically Resistant Indices 700']
  df_train['Visible Atmospherically Resistant Indices RedEdge'] = (b5 - b4) / (b5 + b4)
  # df_train['Weighted Difference Vegetation Index']
  df_train['Wide Dynamic Range Vegetation Index'] = (0.1 * b8 - b4) / (0.1 * b8 + b4)

  #not sure
  df_train['Ashburn Vegetation Index'] = 2 * b9 - b4
  df_train['Canopy Chlorophyll Content Index'] = ((b8 - b5) / (b8 + b5)) / ((b8 - b4) / (b8 + b4))
  df_train['Differenced Vegetation Index MSS'] = 2.4 * b9 - b4
  df_train['Ferric iron, Fe2+'] = b12 / b8 + b3 / b4
  df_train['Ferric iron, Fe3+'] = b4 / b3
  df_train['Ferric Oxides'] = b11 / b8
  df_train['Ferrous iron'] = b12 / b8 + b3 / b4
  df_train['Global Vegetation Moisture Index'] = ((b8 + 0.1) - (b12 + 0.02)) / ((b8 + 0.1) + (b12 + 0.02))
  df_train['Mid-infrared vegetation index'] = b9 / b11
  df_train['Misra Green Vegetation Index'] = -0.386 * b3 - 0.530 * b4 + 0.535 * b6 + 0.532 * b9
  df_train['Misra Non Such Index']         =  0.404 * b3 - 0.039 * b4 - 0.505 * b6 + 0.762 * b9
  df_train['Misra Soil Brightness Index']  =  0.406 * b3 + 0.600 * b4 + 0.645 * b6 + 0.243 * b9
  df_train['Misra Yellow Vegetation Index'] = 0.723 * b3 - 0.597 * b4 + 0.206 * b6 - 0.278 * b9
  df_train['Nonlinear vegetation index'] = (b9 ** 2 - b4) / (b9 ** 2 + b4)
  df_train['Normalized Difference MIR/NIR'] = (b12 - b8) / (b12 + b8)
  df_train['Normalized Difference NIR/MIR'] = (b8 - b12) / (b8 + b12)
  df_train['Normalized Difference NIR/SWIR'] = (b8 - b12) / (b8 + b12)
  df_train['SAVImir'] = (b8 - b12) * (1 + 0.48) / (b8 + b12 + 0.48)
  df_train['Simple Ratio MIR/NIR'] = b12 / b8
  df_train['Simple Ratio MIR/Red'] = b12 / b4
  df_train['Simple Ratio NIR/MIR'] = b8 / b12
  df_train['Soil Background Line'] = b9 -2.4 * b4
  df_train['Soil Composition Index'] = (b11 - b8) / (b11 + b8)
  df_train['Specific Leaf Area Vegetation Index'] = b8 / (b4 + b12)
  # df_train['Tasselled Cap - brightness'] = 0.3037 * b2 + 0.2793 * b3 + 0.4743 * b4 + 0.5585 * b8 + 0.5082 * b10 + 0.1863 * b12
  df_train['Tasselled Cap - Green Vegetation Index MSS'] = -0.283 * b3 - 0.660 * b4 + 0.577 * b6 + 0.388 * b9
  df_train['Tasselled Cap - Non Such Index MSS']         = -0.016 * b3 + 0.131 * b4 - 0.425 * b6 + 0.882 * b9
  df_train['Tasselled Cap - Soil Brightness Index MSS']  =  0.332 * b3 + 0.603 * b4 + 0.675 * b6 + 0.262 * b9
  df_train['Tasselled Cap - vegetation']                 = -0.2848 * b2 - 0.2435 * b3 - 0.5436 * b4 + 0.7243 * b8 + 0.0840 * b11 - 0.1800 * b12
  df_train['Tasselled Cap - wetness']                    = 0.1509 * b2 + 0.1973 * b3 + 0.3279 * b4 + 0.3406 * b8 - 0.7112 * b11 - 0.4572 * b12
  df_train['Tasselled Cap - Yellow Vegetation Index MSS'] = -0.899 * b3 + 0.428 * b4 + 0.076 * b6 - 0.041 * b9
  return df_train
```
